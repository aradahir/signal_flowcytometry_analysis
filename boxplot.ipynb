```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('pos-neg-del-weak(flow).csv')
df['MFI_rHealth']
```

```python
sns.set(style="whitegrid")
fig = plt.figure(figsize =(10,8))
ax = sns.boxplot(x="subtype", y="MFI", data=df, linewidth=3, palette="Set2")
ax = sns.swarmplot(x="subtype", y="MFI", data=df, color=".1")
ax.set_xlabel('subtype', fontsize = 16)
ax.set_ylabel('MFI', fontsize = 16)
ax.set_title('quantitative D typing', fontsize = 20)

fig.savefig("rawdata(flow).pdf", bbox_inches='tight')
```

```python
sns.set(style="whitegrid")
fig = plt.figure(figsize =(10,8))
ax = sns.boxplot(x="subtype", y="MFI_rHealth", data=df, linewidth=3, palette="Set2")
ax = sns.swarmplot(x="subtype", y="MFI_rHealth", data=df, color=".1")
ax.set_xlabel('subtype', fontsize = 16)
ax.set_ylabel('MFI', fontsize = 16)
ax.set_title('quantitative D typing', fontsize = 20)

fig.savefig("rawdata(rHealth).pdf", bbox_inches='tight')
```
