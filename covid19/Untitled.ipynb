```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('D:\covid19\Thai Psychological impact of COVID-19 outbreak on Thai people (Responses).xlsx')
list(data.columns) 
```

```python
df_new = data.copy()

col = ['เพศ','ภูมิภาค','อาชีพแบ่ง','กทม ปริมณฑล อื่นๆ','Facebook','ทางการ','LINE','twitter','youtube','โทรทัศน์', 'หนังสือพิมพ์','คะแนนความรู้','คะแนนการรับรู้','คะแนนความเสี่ยง']
df = df_new[col].reset_index()
df
```

```python
data_dict= {}

for  i in df['index']:
    list_behav = []
    for j in col:
        list_behav.append(df[j][i])
    data_dict[i] = list_behav
    
```

```python
def jaccard(first, second):
    return len(set(first).intersection(second)) / len(set(first)), len(set(first).intersection(second)) 
```

```python
charac = list(data_dict.values())
name = list(data_dict.keys())
```

```python
node = []
vertices = []
share_similarity = []
weight = []

for n in range(0, 754):
    for m in range(0, 754):
        if jaccard(charac[n], charac[m])[0] >= 0.8 and jaccard(charac[n], charac[m])[0] <1:
            node.append(name [n])
            vertices.append(name[m])
            weight.append(jaccard(charac[n], charac[m])[0])
            share_similarity.append(set(charac[n]).intersection(charac[m]))
            
share_similarity
```

```python
import networkx as nx
import networkx.algorithms.bipartite as bipartite

df = pd.DataFrame(node, columns=['name'])
df['vertices'] = vertices
df['weight'] = weight
df

df['name'] = df['name'].astype('object')
df['coef'] = df['vertices'].astype('object')
df['weight'] =df['weight'].astype('object') 
#new_df = df[df.coef.apply(lambda c: c!=[])]
#names = ['name']
#idx = pd.MultiIndex.from_tuples(df[names].values.tolist(), names=names)
#xx = pd.DataFrame(df.coef.tolist(), idx).stack().reset_index(names, name='coef')

#G = nx.convert_matrix.from_pandas_edgelist(df, source= 'name' , target='coef')
G = nx.from_pandas_edgelist(df, 'name' , 'coef', ['weight'])
edges = pd.DataFrame({'source': df['name'] ,
                       'target': df['coef'] ,
                      'weight': df['weight']})
G = nx.from_pandas_edgelist(edges, edge_attr=True)
G.name = 'My Networking'
print(nx.info(G))
```

```python
plt.figure(figsize=(24,16))
elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.8]
emedium = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.6 and d['weight'] <= 0.8]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] <= 0.6]
```
